```python
import datetime
import pytz

import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

from hamsci_psws import grape1
```

```python
mpl.rcParams['font.size']        = 18
mpl.rcParams['font.weight']      = 'bold'
mpl.rcParams['axes.labelweight'] = 'bold'
mpl.rcParams['axes.titleweight'] = 'bold'
mpl.rcParams['axes.grid']        = True
mpl.rcParams['grid.linestyle']   = ':'
mpl.rcParams['figure.figsize']   = np.array([15, 8])
mpl.rcParams['axes.xmargin']     = 0
```

```python
freq            = 10e6 # MHz
sTime           = datetime.datetime(2021,10,28,12, tzinfo=pytz.UTC)
eTime           = datetime.datetime(2021,10,28,18, tzinfo=pytz.UTC)

inventory       = grape1.DataInventory()
inventory.filter(freq=freq,sTime=sTime,eTime=eTime)
grape_nodes     = grape1.GrapeNodes(logged_nodes=inventory.logged_nodes)
```

```python
node_nrs        = inventory.get_nodes()
gds = []
for node in node_nrs:
    print('================================================================================')
    print('Loading Node {!s}'.format(node))
    gd = grape1.Grape1Data(node,freq,sTime,eTime,inventory=inventory,grape_nodes=grape_nodes)
    gd.process_data()
    gds.append(gd)
    print()

mp          = grape1.GrapeMultiplot(gds)
```

```python
# WWV Coordinates
solar_lat   =   40.6683
solar_lon   = -105.0384

color_dct   = {'ckey':'lon'}
xkey        = 'UTC'
events      = [{'datetime':datetime.datetime(2021,10,28,15,35),'label':'X1 Solar Flare'}]
mp.multiplot('filtered',color_dct=color_dct,xkey=xkey,solar_lat=solar_lat,solar_lon=solar_lon,events=events)
plt.show()
```
